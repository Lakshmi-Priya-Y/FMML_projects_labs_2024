#Matplotlib

```python
import requests
url = "https://raw.githubusercontent.com/SP9144/FMML_Data/main/Automobile_data.csv"
response = requests.get(url)
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
from io import StringIO
import matplotlib.pyplot as plt
```

```python
#Loading the dataset
data = StringIO(response.text)
data = pd.read_csv(data)
print("Number of data points = ",len(data))
data.head()
```

```python
#Removing rows which have missing values
data.replace("?",np.nan,inplace = True)
data = data.dropna()
print(len(data))
```

```python
#Dividing the dataset into features(x) and the target variable(y)
x = data.drop("price",axis = 1)
y = pd.to_numeric(data["price"])
```

```python
plt.figure(figsize = (24,7))
plt.scatter(x["make"],y)
plt.xlabel("Make")
plt.ylabel("Price")
plt.title("Car maker vs Price - Scatter Plot")
plt.show()
```

```python
sns.set(rc={'figure.figsize':(24,7)})
sns.boxplot(x = x['make'],y=y,palette = "Set3").set_title("Care Manufacturer vs Price - Box Plot")
```

```python
sns.violinplot(x = x['make'],y=y,palette = "Set3").set_title("Care Manufacturer vs Price - Box Plot")
```

```python
sns.swarmplot(x = x['make'],y=y,palette = "Set3").set_title("Care Manufacturer vs Price - Box Plot")
```

```python
sns.boxplot(x = x['make'],y=pd.to_numeric(x['horsepower']),palette = "Set3").set_title("Care Manufacturer vs Horsepower Box Plot")
```

```python
sns.set(rc={'figure.figsize':(7,5)})
sns.scatterplot(x=pd.to_numeric(x['horsepower']),y=y,color = 'red')
```

```python
sns.jointplot(x=pd.to_numeric(x['horsepower']),y=y,kind="reg",color = 'red')
```

2. **Do you think maximum/minimum values are a clear indication of which body-type is costliest, or does the interpretation improve by also taking the interquartile range into consideration?**

 **Ans.** *While convertibles have the highest maximum value, the interquartile range (IQR) for sedans is larger. The IQR represents the spread of the middle 50% of the data. A larger IQR for sedans suggests a greater range in prices for sedans compared to convertibles.*

3. **Do you think the body-style of a car can be a good predictor for its price?**
